```python
def agg_data(security_code):
    # load the data
    balance_df = pd.read_csv(f'../input/{security_code}_BALANCE.csv')
    income_df = pd.read_csv(f'../input/{security_code}_INCOME.csv')
    cashflow_df = pd.read_csv(f'../input/{security_code}_CASHFLOW.csv')
    summary_df = pd.read_csv(f'../input/{security_code}_SUMMARY.csv')
    
    # aggregate the data
    consolidated_df = pd.merge(balance_df, income_df, 
                           on=['SECURITY_CODE', 
                               'SECURITY_NAME_ABBR', 
                               'REPORT_DATE'], 
                           how='left')
    consolidated_df = pd.merge(consolidated_df, cashflow_df, 
                               on=['SECURITY_CODE', 
                                   'SECURITY_NAME_ABBR', 
                                   'REPORT_DATE'], 
                               how='left')
    consolidated_df = pd.merge(consolidated_df, summary_df, 
                               left_on=['SECURITY_CODE', 
                                   'SECURITY_NAME_ABBR', 
                                   'REPORT_DATE'], 
                               right_on=['SECURITY_CODE', 
                                   'SECURITY_NAME_ABBR', 
                                   'REPORT_DATE'], 
                               how='left')
    
    # rename the columns
    names_dict = {'YSTZ': 'OPERATE_INCOME_YOY', 
              'SJLTZ': 'NETPROFIT_YOY', 
              'BPS': 'BOOK_VALUE_PER_SHARE',
              'MGJYXJJE': 'NETCASH_OPERATE_PER_SHARE', 
              'XSMLL': 'GROSS_MARGIN', 
              'YSHZ': 'OPERATE_INCOME_MOM', 
              'SJLHZ': 'NETPROFIT_MOM', 
              'MONETARYFUNDS_RATIO': 'MONETARYFUNDS_YOY',
              'ACCOUNTS_RECE_RATIO': 'ACCOUNTS_RECE_YOY',
              'INVENTORY_RATIO': 'INVENTORY_YOY',
              'ACCOUNTS_PAYABLE_RATIO': 'ACCOUNTS_PAYABLE_YOY', 
              'ADVANCE_RECEIVABLES_RATIO': 'ADVANCE_RECEIVABLES_YOY', 
              'TOTAL_EQUITY_RATIO': 'TOTAL_EQUITY_YOY',
              'TOTAL_ASSETS_RATIO': 'TOTAL_ASSETS_YOY', 
              'TOTAL_LIAB_RATIO': 'TOTAL_LIAB_YOY',
              'OPERATE_EXPENSE_RATIO': 'OPERATE_EXPENSE_YOY', 
              'TOE_RATIO': 'TOTAL_EXPENSE_YOY', 
              'NETCASH_OPERATE_RATIO': 'NETCASH_OPERATE_YOY', 
              'SALES_SERVICES_RATIO': 'SALES_SERVICES_YOY', 
              'PSC_RATIO': 'PAY_STAFF_CASH_YOY', 
              'NETCASH_INVEST_RATIO': 'NETCASH_INVEST_YOY', 
              'RII_RATIO': 'RECEIVE_INVEST_INCOME_YOY', 
              'CLA_RATIO': 'CONSTRUCT_LONG_ASSET_YOY', 
              'NETCASH_FINANCE_RATIO': 'NETCASH_FINANCE_YOY', 
              'TOTAL_OPERATE_INCOME': 'TOTAL_REVENUE', 
             }
    consolidated_df = consolidated_df.rename(columns=names_dict)
    
    return consolidated_df    
```

```python
# 五粮液 000858
# 格力电器 000651
# 美的集团 000333
        
for security_code in ['000858', '000651', '000333']:
    consolidated_df = agg_data(security_code)
    consolidated_df.to_csv(f'../input/{security_code}_CONSOLIDATED.csv', index=False)
```
